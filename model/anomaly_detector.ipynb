```python
import requests
import pandas as pd

API_KEY = "6515Q1S6ZFGQADKRZADA8G24CD9ANECG6J"
WALLET_ADDRESS = "0x742d35Cc6634C0532925a3b844Bc454e4438f44e"

url = f"https://api.etherscan.io/api?module=account&action=txlist&address={WALLET_ADDRESS}&startblock=0&endblock=99999999&sort=asc&apikey={API_KEY}"

response = requests.get(url).json()
data = response["result"]

df = pd.DataFrame(data)
df.head()
```

###  What This Code Does:
- Connects to the Etherscan API using the provided API key.
- Fetches all Ethereum transactions for the given wallet address.
- Converts the JSON response into a table (DataFrame) using pandas.
- Displays the first 5 transactions as a preview.

The output spreadsheet contains details like:
- hash: Transaction ID
- from / to: Sender and receiver addresses
- value: Amount transferred (in wei)
- timeStamp: Time of transaction
- gas  / gasPrice: Transaction fee details


```python
# Convert JSON data into DataFrame (if not already done)
import pandas as pd

df = pd.DataFrame(data)

# Convert timestamp to readable format
df['timeStamp'] = pd.to_datetime(df['timeStamp'], unit='s')

# Convert value from Wei to ETH (1 ETH = 10^18 Wei)
df['value_ETH'] = df['value'].astype(float) / 10**18

# Keep only useful columns
df_clean = df[['hash', 'from', 'to', 'timeStamp', 'value_ETH', 'gas', 'gasPrice']]

# Save cleaned data
df_clean.to_csv("cleaned_data.csv", index=False)

df_clean.head()
```

This step converts raw Ethereum transaction data into a readable format.  
Timestamps are converted to human-readable dates, values are converted from Wei to ETH,  
and only the key columns are kept for further analysis.


```python
from sklearn.ensemble import IsolationForest

# Select only the ETH value for anomaly detection
X = df_clean[['value_ETH']]

# Train Isolation Forest model
model = IsolationForest(contamination=0.05, random_state=42)  # 5% anomalies
df_clean['anomaly'] = model.fit_predict(X)

# Convert -1  1 (anomaly), 1  0 (normal)
df_clean['anomaly'] = df_clean['anomaly'].map({1: 0, -1: 1})

# Save final results as JSON
df_clean.to_json("anomaly_results.json", orient="records")

df_clean.head()
```

This step uses the Isolation Forest algorithm to detect unusual transactions  
based on their ETH value. Transactions marked as `1` in the `anomaly` column
are considered outliers and will be used for dashboard visualization. transactions marked as `0` are considered normal transactions

###  What is Isolation Forest?
Isolation Forest is an **unsupervised anomaly detection algorithm**.  
It works by **randomly splitting the data into trees** and isolating points that differ significantly.  
Points that get isolated quickly are considered **anomalies (outliers)**.  
This makes it efficient for detecting unusual transactions without needing labeled data.
